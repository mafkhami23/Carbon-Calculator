# **Number of Transportation per Person**

```python
# Create a new DataFrame to store the result
result_df = pd.DataFrame(columns=['Response ID', 'Mode_of_Transportation'])

# Iterate through each row and expand based on non-zero values
for _, row in df.iterrows():
    for col in df.columns[3:-4]:
        count = int(row[col])
        if count > 0:
            temp_df = pd.DataFrame({'Response ID': [row['Response ID']] * count,
                                    'Mode_of_Transportation': [col] * count})
            result_df = pd.concat([result_df, temp_df], ignore_index=True)
```

```python
# Columns to convert from str to int
columns_to_convert = ['Single_Occupancy_Vehicle',
       'Carpooling', 'Public_Transportation', 'EV/Hybrid', 'Electric_Scooter',
       'Walking/Biking', 'Working_from_Home']

# Convert string values to numerical values
for col in columns_to_convert:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')
```

```python
# Convert 'Numeric_Column' to numeric type, coerce non-numeric values to NaN
merged_df['Frequency'] = pd.to_numeric(merged_df['Frequency'], errors='coerce')

# Multiply the two columns
merged_df['Weekly_Emission_[Kg_CO2]'] = (merged_df['Real-World CO2_City (g/mi)'] * merged_df['Distance'] * (merged_df['Single_Occupancy_Vehicle']+merged_df['EV/Hybrid'])) / 1000 +(0.45 * 0.453592 * merged_df['Distance'] * merged_df['Public_Transportation'])

merged_df['Annual_Emission_[Kg_CO2]'] = merged_df['Weekly_Emission_[Kg_CO2]'] * 52
```
